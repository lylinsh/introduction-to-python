```python
import matplotlib.pyplot as plt
import cmath
import numpy as np 
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.arange(-5, 5, 0.5)
y = np.arange(-5, 5, 0.5)
xy = [a+b*1j for a in x for b in y]
z = [ cmath.sqrt(i).real for i in xy]
x, y = np.meshgrid(np.arange(-5, 5, 0.5), np.arange(-5, 5, 0.5))
z = np.array(z)
z = z.reshape((20, 20))
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_title("WireFrame")
ax.set_xlabel("x")
surf = ax.plot_wireframe(x, y, z)
plt.show()
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(x, y, z)
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y = np.sin(x)
z = np.cos(y)
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot(x, y, z)
plt.show()
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.scatter(x, y, z)
plt.show()
```
