```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t  = np.arange(-np.pi, np.pi, np.pi/100)
x = np.cos(t)
print(x)
y = np.sin(t)
print(y)
```

```python
# fig = plt.figure()
plt.title("circle")
plt.xlabel("x=cos(t)")
plt.ylabel("y=sin(t)")
plt.plot(x, y, label="circle", color="#00ff00")
# plt.show()
plt.savefig("./../data/circle.png")
```
